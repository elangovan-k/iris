### To find what values, target holds:

```python
print("target_values \n{}".format(iris_dataset["target"]))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=42) 
# random state is used so that we'll get only one value even if we run the code again and again, the suffled values won't change
```

```python

iris_dataframe=pd.DataFrame(X_train,columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe,c=y_train,figsize=(15,15),marker='o',hist_kwds={'bins':20},s=60,cmap="rainbow")

```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

# Evaluating the Model

```python
y_pred=knn.predict(X_test)
print("Test set predictions \n{}".format(y_pred))
```

```python
print("test score:{:.2f}".format(np.mean(y_pred==y_test)))
```

# Making New predictions:

```python
x_new=np.array([[5,2.9,1,0.2]])
x_new.shape
```
